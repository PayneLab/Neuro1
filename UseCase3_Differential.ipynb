```python
from load_data import load_dataset
from plot_utils import graph_by_variance

from plot_utils import get_altered
import seaborn as sns
```

```python
data = load_dataset()
```

### Variance Graphs
These graphs, like volcano plots, can identify proteins that change sinificantly between the two cell types. Calculating variance reduces the need for data imputation that a p value requires. 

```python
graph_by_variance(data, cell_types = ["Inter","Motor"])
```

### Heatmap of Differential Proteins
A heatmap is another effective method of conveying differences in data. First, we select the proteins that are characteristically different between the samples. Here, that is done based on fold change. Our default cutoff is doubling, which often shows presence versus absence. 

```python
significant_proteins = get_altered(data)
ax = sns.heatmap(significant_proteins, yticklabels=False, label='')
ax.set_title("Log Abundances of Differential Proteins", 
             fontdict={'fontsize':17})
ax = ax.set_ylabel('') 
```
