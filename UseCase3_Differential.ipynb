### Heatmap of Differential Proteins
A heatmap is another effective method of conveying differences in data. First, we select the proteins that are characteristically different between the samples. Here, that is done based on fold ch

```python
significant_proteins = get_altered(data)
ax = sns.heatmap(significant_proteins, yticklabels=False, label='')
ax.set_title("Abundances of Differential Proteins", 
             fontdict={'fontsize':17})
ax = ax.set_ylabel('') 
```
